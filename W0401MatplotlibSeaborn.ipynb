```python
import numpy
mu, sigma = 100, 15
x = mu + sigma * numpy.random.randn(10000)

n, bins, patches = plt.hist(x, 200, density=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,.025,r'$\mu=100,\ \sigma=60$')
plt.axis([40, 160, 0, 0.03])
plt.grid(False)
plt.show()
```

```python
# Working with multiple figures and axes
import numpy
t1 = numpy.arange(0.0, 5.0, 0.1)
t2 = numpy.arange(0.0, 5.0, 0.02)
f=numpy.exp(-t1)*numpy.cos(2*numpy.pi*t1)
f2=numpy.exp(-t2)*numpy.cos(2*numpy.pi*t2)

plt.figure(figsize=(12,5))
plt.subplot(211)
plt.plot(t1, f, 'ro', t2, f2, 'k')

plt.subplot(212)
plt.plot(t2, numpy.cos(2*numpy.pi*t2), 'r--')
#plt.plot(t2, f2, 'r--')

plt.show()
```

```python
# Creating A Time Series Plot With Seaborn

x = np.linspace(0, 15, 31)
data = np.sin(x) + np.random.rand(10, 31) + np.random.randn(10, 1)
df = pd.DataFrame(data).melt()

plt.figure(figsize=(12,5))
sns.lineplot(x="variable", y="value", data=df)
plt.show()
```

```python
# Creating A Time Series Plot With Seaborn

x = np.linspace(0, 20, 40)
data = np.sin(x) + np.random.rand(10, 40) + np.random.randn(10, 1)
df = pd.DataFrame(data).melt()

plt.figure(figsize=(12,5))
sns.lineplot(x="variable", y="value", data=df)
plt.show()
```
